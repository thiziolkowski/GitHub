```python
# MY SQL

CREATE TABLE IF NOT EXISTS `TURMA` (
	`Turma_ID` int NOT NULL AUTO_INCREMENT DEFAULT '',
	`Turma_Nome` int NOT NULL DEFAULT '',
	`Turma_Hora` timestamp NOT NULL DEFAULT '',
	`Turma_DT_Inicio` date NOT NULL DEFAULT '',
	`Turma_DT_Final` date NOT NULL DEFAULT '',
	`Colab_ID` int NOT NULL DEFAULT '',
	`Sala_ID` int NOT NULL DEFAULT '',
	`rDidaticos_ID` int NOT NULL DEFAULT '',
	`Niv_ID` int NOT NULL DEFAULT '',
	`Idioma_ID` int NOT NULL DEFAULT '',
	PRIMARY KEY (`Turma_ID`)
);

CREATE TABLE IF NOT EXISTS `COLABORADORES` (
	`Colab_ID` int NOT NULL AUTO_INCREMENT,
	`Colab_Nome` varchar(50) NOT NULL UNIQUE,
	`Colab_CPF` int NOT NULL UNIQUE,
	`Colab_Funcao` varchar(50) NOT NULL,
	`Colab_Email` varchar(50) NOT NULL UNIQUE,
	`Colab_Fone` varchar(50) NOT NULL,
	PRIMARY KEY (`Colab_ID`)
);

CREATE TABLE IF NOT EXISTS `REC_DIDATICOS` (
	`rDidaticos_ID` int NOT NULL AUTO_INCREMENT,
	`rDidaticos_Nome` varchar(50) NOT NULL UNIQUE,
	`rDidaticos_Tipo` varchar(50) NOT NULL,
	PRIMARY KEY (`rDidaticos_ID`)
);

CREATE TABLE IF NOT EXISTS `NIVEL` (
	`Niv_ID` int NOT NULL AUTO_INCREMENT,
	`Niv_Nome` varchar(50) NOT NULL UNIQUE,
	PRIMARY KEY (`Niv_ID`)
);

CREATE TABLE IF NOT EXISTS `ALUNOS` (
	`Alu_ID` int NOT NULL AUTO_INCREMENT,
	`Alu_Nome` varchar(50) NOT NULL UNIQUE,
	`Alu_Matricula` int NOT NULL UNIQUE,
	`Alu_Mensalidade` decimal(10) NOT NULL,
	`Turma_ID` int NOT NULL,
	PRIMARY KEY (`Alu_ID`)
);

CREATE TABLE IF NOT EXISTS `SALA` (
	`Sala_ID` int NOT NULL AUTO_INCREMENT,
	`Sala_Nome` varchar(50) NOT NULL UNIQUE,
	`Sala_Capacidade` int NOT NULL,
	PRIMARY KEY (`Sala_ID`)
);

CREATE TABLE IF NOT EXISTS `PROFESSORES` (
	`Prof_ID` int NOT NULL AUTO_INCREMENT,
	`Prof_Nome` varchar(50) NOT NULL UNIQUE,
	`Prof_CPF` int NOT NULL UNIQUE,
	`Prof_Especializacao` varchar(50) NOT NULL,
	`Colab_Email` varchar(50) NOT NULL UNIQUE,
	`Colab_Fone` varchar(50) NOT NULL,
	`Idioma_ID` int NOT NULL,
	PRIMARY KEY (`Prof_ID`)
);

CREATE TABLE IF NOT EXISTS `IDIOMA` (
	`Idioma_ID` int NOT NULL AUTO_INCREMENT,
	`Idioma_Nome` varchar(50) NOT NULL,
	PRIMARY KEY (`Idioma_ID`)
);

ALTER TABLE `TURMA` ADD CONSTRAINT `TURMA_fk5` FOREIGN KEY (`Colab_ID`) REFERENCES `COLABORADORES`(`Colab_ID`);

ALTER TABLE `TURMA` ADD CONSTRAINT `TURMA_fk6` FOREIGN KEY (`Sala_ID`) REFERENCES `SALA`(`Sala_ID`);

ALTER TABLE `TURMA` ADD CONSTRAINT `TURMA_fk7` FOREIGN KEY (`rDidaticos_ID`) REFERENCES `REC_DIDATICOS`(`rDidaticos_ID`);

ALTER TABLE `TURMA` ADD CONSTRAINT `TURMA_fk8` FOREIGN KEY (`Niv_ID`) REFERENCES `NIVEL`(`Niv_ID`);

ALTER TABLE `TURMA` ADD CONSTRAINT `TURMA_fk9` FOREIGN KEY (`Idioma_ID`) REFERENCES `IDIOMA`(`Idioma_ID`);



ALTER TABLE `ALUNOS` ADD CONSTRAINT `ALUNOS_fk4` FOREIGN KEY (`Turma_ID`) REFERENCES `TURMA`(`Turma_ID`);

ALTER TABLE `PROFESSORES` ADD CONSTRAINT `PROFESSORES_fk6` FOREIGN KEY (`Idioma_ID`) REFERENCES `IDIOMA`(`Idioma_ID`);
```

# ***WEBSCRAPING RECEITA*** - CONSULTA CNPJ

```python
from bs4 import BeautifulSoup
import requests

cnpj = (input("Informe o CNPJ para consulta: "))

# Gerando URL de busca para API
url = f"https://www.receitaws.com.br/v1/cnpj/{cnpj}"

# Verifica se teve retorno
html = requests.get(url)
if html.status_code == 200:
    soup = BeautifulSoup(html.content, 'html.parser')
    print(soup.prettify())
else:
    print(f"Error: Request failed with status code {html.status_code}")




```
